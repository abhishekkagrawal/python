### 4. Data Normalization

```python
def capitalize(line):
    return ' '.join(s[:1].upper() + s[1:] for s in line.split(' '))

def truncateSecondWord(words):
    word_list = words.split()
    if len(word_list)==2 :
        return(capitalize(' '.join([word_list[0],word_list[1][0]])))
    return capitalize(word_list[0])
    
df_olympics["Discipline"] = df_olympics["Discipline"].apply(truncateSecondWord)

print(df_olympics["Discipline"].value_counts().sort_index(0))
print("Distinct Discipline of Sports:", set(df_olympics["Discipline"]))                    
```

```python
df_olympics.insert(0, 'ID', range(1, 1 + len(df_olympics)))
df_olympics.head()
```

- ### Check for Null Values

```python
df_country_mode = df_olympics[(df_olympics["City"]=="London") & (df_olympics["Gender"].isin(["Women","Men"]))
                  & (df_olympics["Year"]==2012) 
                & (df_olympics['Country'].notnull())]['Country'].mode()[0]
df_olympics['Country'] = df_olympics['Country'].fillna(df_country_mode)

df_olympics["Country"].value_counts().head(50)
```


Countries with most number of medals won in all Olympics.

```python
groupcountry = df_olympics.groupby(by=['Country'],as_index= False).sum()
top50 = groupcountry.sort_values(by=['allmedals'],ascending = False).head(10)
plot2 = sns.barplot('Country','allmedals',data=top50).set_xticklabels(top50.Country,rotation=82)
```

- ### The Sport and the Country that dominated.


```python
groupsport = df_olympics.groupby(by=['Sport','Country'],as_index=False).sum()
spotop = pd.DataFrame()
sp = df_olympics.Sport.unique()
for i in sp:
    spotop = pd.concat([spotop,groupsport[groupsport['Sport'] == i].sort_values(by=['allmedals'],ascending= False).head(1)])
spotop = spotop[['Sport','Country','Bronze','Gold','Silver','allmedals']]
spotop.sort_values('allmedals',ascending=False)
```
